```python
import pandas as pd

def split_text_preserving_sentences(essay, labels, chunk_size=500):
    # This function attempts to split the text into chunks that are close to chunk_size words,
    # but without splitting in the middle of a sentence.
    words = essay.split()
    current_chunk = []
    label_chunks = []
    word_chunks = []
    current_label_chunk = []
    word_count = 0

    for i, word in enumerate(words):
        current_chunk.append(word)
        current_label_chunk.append(labels[i])
        word_count += 1

        # Check if we're at a potential split point
        if word_count >= chunk_size and (word.endswith('.') or word.endswith('!') or word.endswith('?')):
            word_chunks.append(' '.join(current_chunk))
            label_chunks.append(current_label_chunk)
            current_chunk = []
            current_label_chunk = []
            word_count = 0

    # Add the last chunk if there is any residue
    if current_chunk:
        word_chunks.append(' '.join(current_chunk))
        label_chunks.append(current_label_chunk)

    return word_chunks, label_chunks

# Assuming `grouped` is your DataFrame and it contains 'essay' and 'labels' columns
new_rows = []

for _, row in grouped.iterrows():
    essay_chunks, label_chunks = split_text_preserving_sentences(row['essay'], row['labels'], chunk_size=500)
    for chunk, label_chunk in zip(essay_chunks, label_chunks):
        new_rows.append({'essay': chunk, 'labels': label_chunk})

new_df = pd.DataFrame(new_rows)

# Now new_df contains the essays split close to 500 words, attempting not to split sentences in the middle
```

```python
def sanity_check(df):
    for index, row in df.iterrows():
        # Split the essay into tokens
        tokens = row['essay'].split()
        # Retrieve the labels
        labels = row['labels']

        # Check if the number of tokens matches the number of labels
        if len(tokens) != len(labels):
            print(f"Row {index} has a mismatch - {len(tokens)} tokens and {len(labels)} labels")
            return False  # Return False if any mismatch is found
    return True  # Return True if all rows pass the sanity check

# Assuming new_df is your new DataFrame with the split essays and corresponding labels
if sanity_check(new_df):
    print("Sanity check passed: Each token has a corresponding label.")
else:
    print("Sanity check failed: There's a mismatch between tokens and labels.")
```
