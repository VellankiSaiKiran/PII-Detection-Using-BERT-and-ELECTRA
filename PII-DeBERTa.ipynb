```python
# Correcting the code to handle the situation where the column items are not strings

def preprocess_column(column):
    # Convert the string representation of a list into an actual list
    # It involves stripping the brackets and then splitting by comma and stripping extra spaces.
    processed_column = []
    for item in column:
        # Ensure the item is a string
        item = str(item)
        # Strip the square brackets and split the string into a list of tokens
        tokens = item.strip('[]').split(', ')
        # Remove any possible surrounding quotes and extra spaces from each token
        processed_column.append([token.strip("\"' ") for token in tokens])
    return processed_column

sentences = preprocess_column(df['token'])
labels = preprocess_column(df['label'])
```

```python
# Convert labels to numerical IDs
unique_labels = set(lbl for doc in labels for lbl in doc)
label2id = {label: i for i, label in enumerate(unique_labels)}
id2label = {i: label for label, i in label2id.items()}
labels = [[label2id[lbl] for lbl in doc] for doc in labels]
```

```python
from transformers import DebertaTokenizerFast

# Instantiate the tokenizer with the correct configuration for pre-tokenized inputs
tokenizer = DebertaTokenizerFast.from_pretrained('microsoft/deberta-base', add_prefix_space=True)

ignore_label = -100
def align_labels_with_tokens(encodings, labels_per_sentence):
    aligned_labels = []
    for i, label in enumerate(labels_per_sentence):
        word_ids = encodings.word_ids(batch_index=i)  # Get the word_ids for the i-th sentence
        aligned_label = []
        previous_word_id = None
        for word_id in word_ids:
            if word_id is None:
                # Special tokens have no associated word in the original sentence
                aligned_label.append(ignore_label)
            elif word_id != previous_word_id:
                # Only the first token of a given word will use the label
                aligned_label.append(label[word_id])
                previous_word_id = word_id
            else:
                # For the remaining sub-tokens of a given word, set the label to -100
                aligned_label.append(ignore_label)
        aligned_labels.append(aligned_label)
    return aligned_labels

# Apply the alignment function to your tokenized data
train_encodings = tokenizer(X_train, is_split_into_words=True, return_tensors="pt", padding=True, truncation=True, max_length=512)
val_encodings = tokenizer(X_val, is_split_into_words=True, return_tensors="pt", padding=True, truncation=True, max_length=512)

train_aligned_labels = align_labels_with_tokens(train_encodings, y_train)
val_aligned_labels = align_labels_with_tokens(val_encodings, y_val)


```

```python
# Assuming `train_aligned_labels` and `val_aligned_labels` are the output from the previous step

# You might need to pad the labels if they are not all the same length
def pad_labels(labels, max_len, pad_value=-100):
    return [label + [pad_value] * (max_len - len(label)) for label in labels]

# Find the maximum length to pad all sequences to this length for consistency
max_length = max(max(len(lbl) for lbl in train_aligned_labels), max(len(lbl) for lbl in val_aligned_labels))

# Pad the labels
train_labels_padded = pad_labels(train_aligned_labels, max_length)
val_labels_padded = pad_labels(val_aligned_labels, max_length)

# Convert the padded labels to tensors
train_labels_tensor = torch.tensor(train_labels_padded)
val_labels_tensor = torch.tensor(val_labels_padded)
```

```python
class TokenClassificationDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        # `__getitem__` should return a dictionary with the input_ids, attention_mask, and labels
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = self.labels[idx]
        return item

    def __len__(self):
        return len(self.labels)

# Instantiate the custom datasets
train_dataset = TokenClassificationDataset(train_encodings, train_labels_tensor)
val_dataset = TokenClassificationDataset(val_encodings, val_labels_tensor)

# Create the DataLoader for both sets
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)
```

```python

# Load the pre-trained DeBERTa model
model = DebertaForTokenClassification.from_pretrained('microsoft/deberta-base', num_labels=len(unique_labels))

# Define training arguments
training_args = TrainingArguments(
    output_dir='./results',           # output directory
    num_train_epochs=3,               # total number of training epochs
    per_device_train_batch_size=8,   # batch size per device during training
    per_device_eval_batch_size=8,    # batch size for evaluation
    warmup_steps=500,                 # number of warmup steps for learning rate scheduler
    weight_decay=0.01,                # strength of weight decay
    logging_dir='./logs',             # directory for storing logs
    logging_steps=10,
    evaluation_strategy="epoch",      # evaluation is done at the end of each epoch
)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
)

# Train the model
trainer.train()

# Evaluate the model
results = trainer.evaluate()

# Save the model
model.save_pretrained("/content/drive/My Drive/ML-Project/PII-DATA/deberta_classifier")

# Print the evaluation results
print(results)
```
